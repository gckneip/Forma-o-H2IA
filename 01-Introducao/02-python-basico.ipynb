# Python Básico para IA

Este notebook revisa conceitos fundamentais de Python necessários para trabalhar com Inteligência Artificial.

## 1. Estruturas de Dados Básicas

```python
# Listas
numeros = [1, 2, 3, 4, 5]
print("Lista:", numeros)

# Dicionários
dados = {'nome': 'João', 'idade': 25, 'curso': 'IA'}
print("Dicionário:", dados)

# Tuplas
coordenadas = (10, 20)
print("Tupla:", coordenadas)
```

## 2. List Comprehensions

```python
# Criando uma lista de quadrados
quadrados = [x**2 for x in range(10)]
print("Quadrados:", quadrados)

# Filtrando números pares
pares = [x for x in range(20) if x % 2 == 0]
print("Pares:", pares)
```

## 3. Funções

```python
def calcular_media(valores):
    """Calcula a média de uma lista de valores."""
    if not valores:
        return 0
    return sum(valores) / len(valores)

# Testando a função
notas = [7.5, 8.0, 9.0, 6.5]
media = calcular_media(notas)
print(f"Média das notas: {media:.2f}")
```

## 4. Funções Lambda

```python
# Função lambda para elevar ao quadrado
quadrado = lambda x: x**2
print("5 ao quadrado:", quadrado(5))

# Usando lambda com map
numeros = [1, 2, 3, 4, 5]
dobro = list(map(lambda x: x * 2, numeros))
print("Dobro:", dobro)
```

## 5. Classes e Objetos

```python
class ModeloIA:
    """Classe base para modelos de IA."""
    
    def __init__(self, nome):
        self.nome = nome
        self.treinado = False
    
    def treinar(self, dados):
        """Treina o modelo com os dados fornecidos."""
        print(f"Treinando modelo {self.nome}...")
        self.treinado = True
    
    def prever(self, entrada):
        """Faz previsões com o modelo."""
        if not self.treinado:
            return "Modelo precisa ser treinado primeiro!"
        return f"Previsão para {entrada}"

# Usando a classe
modelo = ModeloIA("Classificador")
modelo.treinar([1, 2, 3])
print(modelo.prever("nova_entrada"))
```
